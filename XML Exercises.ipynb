```python
from xml.etree import ElementTree as ET
import pandas as pd
```

```python
#read in data
document_tree = ET.parse("./data/mondial_database.xml",)
document_root = document_tree.getroot()
```

```python
#1 10 countries with the lowest infant mortality rate
```

```python
#creating empty df for country mortality
df_country_mort=pd.DataFrame(columns=['Country', "Infant_Mortality"])
```

```python
for country in document_root.findall('country'):
    country_name=country.find("name").text
    if country.find("infant_mortality") is None:
        continue;
    country_mort=country.find('infant_mortality').text
    df_country_mort=df_country_mort.append(
    {"Country": country_name,
    "Infant_Mortality": country_mort}, ignore_index=True)
```

```python
#changing infant mortality rate to float
df_country_mort['Infant_Mortality']=df_country_mort.Infant_Mortality.astype(float)
```

```python
df_country_mort.sort_values(by="Infant_Mortality", ascending=True).head(10)
```

```python
#2  10 Cities with the largest population
df_city_pop=pd.DataFrame(columns=["city", "year", "population"])
```

```python
for country in document_root.getiterator("country"):
    for city in country.iter("city"):
        city_name=city.find("name").text
        
        if city.findall("population") is None:
            continue;
            
        for pop in city.findall("population"):
            year=pop.attrib['year']
            population=pop.text
            #print(population)
            
            df_city_pop = df_city_pop.append(
            {
                "city": city_name,
                "year": year,
                "population": population
            }, ignore_index = True)
```

```python
df_city_pop['population']=df_city_pop.population.astype(float)
```

```python
df_city_pop.sort_values(by="population", ascending=False).head(10)
```

```python
df_city_pop.info()
```

```python
#3 10 ethinic groups with the largest overal populations(sum of best/latest estimates over all countries)
```

```python
#creating empty df for country populations by ethnic group
df_country_eg=pd.DataFrame(columns=["country", "ethnic group", "population"])
```

```python
for country in document_root.findall('country'):
    country_name=country.find("name").text
    country_pop=country.find("population").text
    if country.find("ethnicgroup") is None:
        continue;
    for pop in country.findall("ethnicgroup"):
        percentage_eg=pop.attrib['percentage']
        number_eg=(float(percentage_eg)/100)*float(country_pop)
        country_eg=pop.text

        df_country_eg= df_country_eg.append(
        {"country":country_name,
        "ethnic group": country_eg,
        "population":number_eg}, ignore_index=True)
df_country_eg.head(20)
```

```python
#changing population to float
#df_country_eg['population']=df_country_eg.population.astype(float)
```

```python
#summing popluation by ethnic group
eg_totals=df_country_eg.groupby('ethnic group').sum()
eg_totals
```

```python
#10 ethnic groups with highest population
eg_totals.sort_values(by="population", ascending=False ).head(10)
```

```python
#4 Name and country of a) longest river, b) largest lake and c) airport at highest elevation
```

```python
#creating df for country code + name
df_country_code=pd.DataFrame(columns=['country_name', 'country_code'])
```

```python
#populating df for country code + name
for country in document_root.findall('country'):
    country_name=country.find("name").text
    country_code=country.attrib['car_code']
   
    df_country_code=df_country_code.append(
    {"country_name": country_name,
    "country_code": country_code}, ignore_index=True)
```

```python
df_country_code
```

```python
#creating empty df for country populations by ethnic group
df_river=pd.DataFrame(columns=["country_code", "river name", "river_length"])
```

```python
#populating dataframe from xml 
for river in document_root.findall('river'):
    if river.find('length') is None:
        continue;
    river_country=river.attrib['country']
    river_name=river.find("name").text
    river_length=river.find("length").text
    
    df_river=df_river.append(
    {"country_code": river_country, "river name": river_name, "river_length":river_length}, ignore_index=True)
```

```python
#changin river length from string to float
df_river['river_length']=df_river.river_length.astype(float)
```

```python
#merging in the country names
df_river=pd.merge(df_river, df_country_code, on="country_code")
```

```python
#Country and name of longest river
df_river.sort_values(by='river_length', ascending=False).head(1)
```

```python
#creating empty DataFrame for Lake info
df_lake=pd.DataFrame(columns=["country_code", "name", "area"])
```

```python
#populating dataframe from xml 
for lake in document_root.findall('lake'):
    if lake.find('area') is None:
        continue;
    lake_country=lake.attrib['country']
    lake_name=lake.find("name").text
    lake_area=lake.find("area").text
    
    df_lake=df_lake.append(
    {"country_code": lake_country, "name": lake_name, "area":lake_area}, ignore_index=True)
```

```python
#changing area from string to float
df_lake['area']=df_lake.area.astype(float)
```

```python
#merging in the country names
df_lake=pd.merge(df_lake, df_country_code, on="country_code")
```

```python
# country and name of largest lake
df_lake.sort_values(by="area", ascending=False).head(1)
```

```python
#creating empty DataFrame for airport info
df_airport=pd.DataFrame(columns=["country_code", "airport", "elevation"])
```

```python
#populating dataframe from xml 
for airport in document_root.findall('airport'):
    if airport.find('elevation') is None:
        continue;
    airport_country=airport.attrib['country']
    airport_name=airport.find("name").text
    airport_elevation=airport.find("elevation").text
    
    df_airport=df_airport.append(
    {"country_code": airport_country, "airport": airport_name, "elevation":airport_elevation}, ignore_index=True)
```

```python
#changing elevation from string to float
df_airport['elevation']=df_airport.elevation.astype(float)
```

```python
#merging in the country names
df_airport=pd.merge(df_airport, df_country_code, on="country_code")
```

```python
# country and name of airport at highest elevation
df_airport.sort_values(by="elevation", ascending=False).head(1)
```
